```python
import numpy as np
from matplotlib import pyplot as plt
from scipy.linalg import sqrtm
import scipy as sp

############### Generate the samples using given mean and covariance #########
d = 2
n = 1000
X,X2 = [],[]
mu1 = np.array([-1,1])
sigma1 = np.array([[1,0],[0,2]])
mu2 = np.array([1,1])
sigma2 = np.array([[1,0],[0,2]])
x1,x2 = [],[]
A = sqrtm(sigma1)
A2 = sqrtm(sigma2) 
b = mu1
for i in range(n):
  X.append(np.random.randn(d))
  X2.append(np.random.randn(d))
  X[i] = np.dot(A,X[i]) + mu1
  X2[i] = np.dot(A2,X2[i]) + mu2
  Z = X[i]
  x1.append(Z[0])
  x2.append(Z[1])


################### Join the 2 samples together using concatenate ###########
X = np.array(X)
X2 = np.array(X2)
X = np.concatenate((X,X2),axis = 0)
Y = np.array([1]*n)
Y2 = np.array([-1]*n)
Y = np.concatenate((Y,Y2),axis = 0).reshape(2*n,1)
lambd = 0.1
# Y = np.reshape(10000,1)


d = X.shape[1]
B = 0
n = X.shape[0]



bb = 0
b = 0
eta = 0.1
t = 1
WW = []
BB = []
s = np.array([[0],[0]])
# print(np.shape(w))


################## Check 100 times #########################3
while t <= 100:
  
  w = np.random.rand(2,1)



  # print(f.shape)
  # print(Y.shape)

  #Gradient for w
  #frst = (0 + np.dot(X,w))

################### Implement GD for each weight generated ###################
  converge = 0
  while converge != d:
      w_old = w


      # GD for weight vector
      O = 1/(1+np.exp(-Y * (B + np.dot(X,w))))
      GG = (X*((O-1)*Y))
      GG_1 = np.sum(GG,axis=0).reshape(2,1)/n
      w = w - eta * GG_1


      # GD for b value
      m1 = 1/(1+np.exp(-Y * (B + np.dot(X,w))))
      B = B - eta * (m1-1) * Y/n
      b = np.mean(B)
      # cost_func = 1/n * np.sum(np.log(1+np.exp(np.multiply(-Y,B+np.dot(w.T,X.T))))) + lambd* np.sum(pow(w,2))

      
      converge = 0
      for i in range(len(w)):
        if np.absolute(w_old[i] - w[i]) <= 0.0001:
          converge = converge + 1
     
  WW.append(w)
  s = s+w


  bb = bb + b
  t=t+1
  #print(t)

WER = s/100

b = b/100


################### Accuracy for classifier learnt using GD #################
y_hat = []
sig = 1/(1+np.exp(-np.dot(X,WER))) + b

for i in range(n):
  if sig[i] < 0.5:
    y_hat.append(-1)
  elif sig[i] > 0.5:
    y_hat.append(1)

acc = 0

for i in range(n):
  if y_hat[i] == Y[i]:
    acc = acc + 1
print("WERM Accuracy")
print(acc*100/n)


######################## Accuracy  for Bayes classifier ###############
W_bayes  = np.array([[-4],[0]])
b = 0
y_hat_1 = []
sig1 = 1/(1+np.exp(-np.dot(X,W_bayes))) + b

for i in range(n):
  if sig1[i] < 0.5:
    y_hat_1.append(-1)
  elif sig1[i] > 0.5:
    y_hat_1.append(1)

acc1 = 0

for i in range(n):
  if y_hat_1[i] == Y[i]:
    acc1 = acc1 + 1
print("W Bayes accuracy")
print(acc1*100/n)

```
